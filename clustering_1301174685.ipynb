### Data Splitting

Membagi dataset menjadi dua bagian yaitu Training Dataset dan Testing Dataset

```
dataset_train = data[:2500]
dataset_test = data[2500:]
dataset = data

training_set = dataset_train.iloc[:, 0:2].values
dataset_train.head(10)
```

```
import numpy as np
Min_Att1 = np.min(training_set[:, 0])
Max_Att1 = np.max(training_set[:, 0])
Min_Att2 = np.min(training_set[:, 1])
Max_Att2 = np.max(training_set[:, 1])
print(Min_Att1)
print(Max_Att1)
print(Min_Att2)
print(Max_Att2)
```

```
np.random.seed(200)
k = 4
centroids = {
    i+1: [np.random.uniform(Min_Att1, Max_Att1), np.random.uniform(Min_Att2, Max_Att2)]
    for i in range(k)
}
colmap = {1: 'red', 2: 'green', 3: 'blue', 4:'yellow', 5:'black'}
print(centroids)
```

```
def k_means(df, centroids):
    for i in centroids.keys():
        df['distance_from_{}'.format(i)] = (
            np.sqrt(
                (df['price'] - centroids[i][0]) ** 2
                + (df['odometer'] - centroids[i][1]) ** 2 
            )
        )
    centroid_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()]
    df['closest'] = df.loc[:, centroid_distance_cols].idxmin(axis=1)
    df['closest'] = df['closest'].map(lambda x: int(x.lstrip('distance_from_')))
    df['color'] = df['closest'].map(lambda x: colmap[x])
    return df
```

```
import copy
def update(k):
    for i in centroids.keys():
        centroids[i][0] = np.mean(df[df['closest'] == i]['price'])
        centroids[i][1] = np.mean(df[df['closest'] == i]['odometer'])
    return k 
```

```
np.random.seed(200)
data = dataset
df = data
k = 4
centroids = {
    i+1: [np.random.uniform(Min_Att1, Max_Att1), np.random.uniform(Min_Att2, Max_Att2)]
    for i in range(k)
}
colmap = {1: 'red', 2: 'green', 3: 'blue', 4:'yellow'}
print(centroids)

df = k_means(data, centroids)
while True:
    closest_centroids = df['closest'].copy(deep=True)
    centroids = update(centroids)
    df = k_means(df, centroids)
    if closest_centroids.equals(df['closest']):
        break
df.tail(25)
```
